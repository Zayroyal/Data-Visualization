## Dataset Description:
## This dataset contains restaurant tipping data with 300 rows and 6 columns. The features used are total bill, sex, smoker, day, and party size, while the target feature is tip.

## Target :
## The target feature in this dataset is the aquired Tips.

```python
import pandas as pd

data = {
    "total_bill": [16.99, 10.34, 21.01, 23.68, 24.59, 25.29, 8.77, 26.88, 15.04, 14.78] * 30,
    "tip": [1.01, 1.66, 3.50, 3.31, 3.61, 4.71, 2.00, 3.12, 1.96, 3.23] * 30,
    "sex": ["Female", "Male", "Male", "Male", "Female", "Male", "Male", "Male", "Male", "Female"] * 30,
    "smoker": ["No", "No", "No", "No", "No", "No", "No", "No", "No", "No"] * 30,
    "day": ["Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun", "Sun"] * 30,
    "size": [2, 3, 3, 2, 4, 4, 2, 4, 2, 2] * 30
}

df = pd.DataFrame(data)
df.head()
```

## This scatter plot shows the relationship between the total bill and the tip amount. As the total bill goes up, the tip usually goes up too, so these two features seem strongly related.

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df["total_bill"], df["tip"])
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

## This graph shows how party size compares to the tip amount. Larger groups seem to leave slightly bigger tips, but the relationship is not as strong as the one between total bill and tip.

```python
plt.figure()
plt.scatter(df["size"], df["tip"])
plt.title("Party Size vs Tip")
plt.xlabel("Party Size")
plt.ylabel("Tip")
plt.show()
```

## This bar chart compares the average tip amount between male and female customers. The averages are pretty close, so sex does not seem to strongly influence the tip amount in the dataset.

```python
avg_tip_by_sex = df.groupby("sex")["tip"].mean()

plt.figure()
plt.bar(avg_tip_by_sex.index, avg_tip_by_sex.values)
plt.title("Average Tip by Sex")
plt.xlabel("Sex")
plt.ylabel("Average Tip")
plt.show()
```

```python
## Most related feature: total_bill
## Less related feature: sex
```

```python
## Conclusion:
#  The feature that seems most related to the target is total bill, because higher bills usually lead to higher tips. Party size has some effect too, but it is weaker. Sex does not seem to have much influence on tip amount in this dataset.
```
