```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("spotify-2023.csv", encoding="ISO-8859-1")

df.head()
```

```python
plt.hist(df["bpm"], bins=30)
plt.title("Distribution of BPM")
plt.xlabel("BPM")
plt.ylabel("Number of Songs")
plt.show()
```

```python
df["key"].value_counts().plot(kind="bar")
plt.title("Distribution of Musical Keys")
plt.xlabel("Key")
plt.ylabel("Number of Songs")
plt.show()
```

```python
df["mode"].value_counts().plot(kind="bar")
plt.title("Distribution of Mode")
plt.xlabel("Mode")
plt.ylabel("Number of Songs")
plt.show()
```

```python
plt.hist(df["danceability_%"], bins=30)
plt.title("Distribution of Danceability")
plt.xlabel("Danceability (%)")
plt.ylabel("Number of Songs")
plt.show()
```

```python
plt.hist(df["energy_%"], bins=30)
plt.title("Distribution of Energy")
plt.xlabel("Energy (%)")
plt.ylabel("Number of Songs")
plt.show()
```

```python
plt.hist(df["speechiness_%"], bins=30)
plt.title("Distribution of Speechiness")
plt.xlabel("Speechiness (%)")
plt.ylabel("Number of Songs")
plt.show()
```

```python
plt.hist(df["acousticness_%"], bins=30)
plt.title("Distribution of Acousticness")
plt.xlabel("Acousticness (%)")
plt.ylabel("Number of Songs")
plt.show()
```

```python
hits = df[
    (df["bpm"].between(90,130)) &
    (df["danceability_%"].between(60,90)) &
    (df["energy_%"].between(60,85)) &
    (df["speechiness_%"] < 20) &
    (df["acousticness_%"] < 50)
]

hits[["track_name","artist(s)_name"]]
```
