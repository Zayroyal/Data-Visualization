```python
##This bar chart shows how many people prefer each type of fruit based on the data. From this, you can see that strawberries and bananas are the most popular, while oranges are the least preferred, showing a clear difference in taste preferences.
```

```python
import matplotlib.pyplot as plt

fruits = ["Apples", "Bananas", "Oranges", "Grapes", "Strawberries"]
preferences = [15, 25, 10, 20, 30]

plt.figure()
plt.bar(fruits, preferences)
plt.title("Fruit Preference Survey")
plt.xlabel("Fruit Type")
plt.ylabel("Number of People")
plt.show()
```

```python
## This pie chart shows how each fruit makes up a portion of the total preferences. It’s clear that strawberries take up the biggest percentage, while oranges take up the smallest, showing which fruits people favor the most overall.
```

```python
import matplotlib.pyplot as plt

fruits = ["Apples", "Bananas", "Oranges", "Grapes", "Strawberries"]
preferences = [15, 25, 10, 20, 30]

plt.figure()
plt.pie(preferences, labels=fruits, autopct='%1.1f%%')
plt.title("Fruit Preference Distribution")
plt.show()
```

```python
##This horizontal bar chart shows the same fruit preferences but in a way that’s easier to compare side by side. It makes it more obvious that strawberries and bananas are ahead of the others, while oranges fall behind in popularity.
```

```python
import matplotlib.pyplot as plt

fruits = ["Apples", "Bananas", "Oranges", "Grapes", "Strawberries"]
preferences = [15, 25, 10, 20, 30]

plt.figure()
plt.barh(fruits, preferences)
plt.title("Fruit Preference Comparison")
plt.xlabel("Number of People")
plt.ylabel("Fruit Type")
plt.show()
```
